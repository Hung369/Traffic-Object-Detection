```python
from ultralytics import YOLO
import locale
locale.getpreferredencoding = lambda: "UTF-8" #convert locale to UTF8 format
from IPython.display import HTML
from base64 import b64encode
```

```python
# Load a model
model = YOLO('yolov8n.pt')

# Train model
results = model.train(data="/content/drive/MyDrive/Training_YOLO/addition/data.yaml", epochs=5, imgsz=640, device=0, save=True)
```

```python
model = YOLO('/content/drive/MyDrive/Training_YOLO/runs/detect/train/weights/last.pt') # continue training on addition dataset
results = model.train(data="/content/drive/MyDrive/Training_YOLO/addition/data.yaml", epochs=10, imgsz=640, device=0, save=True)
```

```python
model = YOLO('/content/drive/MyDrive/Training_YOLO/runs/detect/train2/weights/last.pt') # continue training on addition dataset
results = model.train(data="/content/drive/MyDrive/Training_YOLO/addition/data.yaml", epochs=3, imgsz=640, device=0, save=True)
```
